```python
# Importing required libraries
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib as mpl
import matplotlib.pyplot as plt
# import warnings
# warnings.filterwarnings('ignore')
```

```python
# import data about world happiness in 2019 
# datasets are csv files downloaded from https://www.kaggle.com/unsdsn/world-happiness
data2019 = pd.read_csv("2019.csv")
data2018 = pd.read_csv("2018.csv")
data2017 = pd.read_csv("2017.csv")
data2016 = pd.read_csv("2016.csv")
data2015 = pd.read_csv("2015.csv")

print("Top 5 Happiest Country in 2018")
print(data2018.head())
print()

print("Top 5 Happiest Country in 2019")
print(data2019.head())
```

# Guiding Question 1: Does a greater GDP/Economy correlate to a higher happiness score?

**Hypothesis 1: I hypothesize that the more prosperous a country is (economy/GDP), the higher the happiness score.**

```python
fig, axes = plt.subplots(3, 2, sharex = True, sharey = True, figsize = (20, 20))

# Plot 5 subplots (scatter plots) for year 2015-2019
axes[0, 0].scatter(data2015['Economy (GDP per Capita)'], data2015['Happiness Score'], marker ='o', color = 'c', label='2015', s = 20)
axes[0, 0].set_title('Year 2015', size = 15)

axes[0, 1].scatter(data2016['Economy (GDP per Capita)'], data2016['Happiness Score'], marker ='o', color = 'y', label='2016', s = 20)
axes[0, 1].set_title('Year 2016', size = 15)

axes[1, 0].scatter(data2017['Economy..GDP.per.Capita.'], data2017['Happiness.Score'], marker ='o', color = 'g', label='2017', s = 20)
axes[1, 0].set_title('Year 2017', size = 15)

axes[1, 1].scatter(data2018['GDP per capita'], data2018['Score'], marker ='o', color = 'm', label='2018', s = 20)
axes[1, 1].set_title('Year 2018', size = 15)

axes[2, 0].scatter(data2019['GDP per capita'], data2019['Score'], marker ='o', color = 'b', label='2019', s = 20)
axes[2, 0].set_title('Year 2019', size = 15)

axes[2, 1].set_visible(False)

fig.text(.5, .1, 'Economy (GDP per capita)', ha='center', fontsize = 20)
fig.text(.1, .5, 'Happiness Score', ha='center', rotation='vertical', fontsize = 20)

plt.show()
```

# Guiding Question 2: Do all the factors in the original datasets correlate to happiness score? If so, which one has the strongest relationship? (In 2019)

**Hypothesis 2: I hypothesize that GDP, social support, healthy life expectancy, freedom to make life choices, and generosity are positively correlated with happiness score in 2019.**

```python
fig, axes = plt.subplots(3, 2, sharey = True, figsize = (20, 20))

axes[0, 0].scatter(data2019['GDP per capita'], data2019['Score'], marker ='o', color = 'b', s = 20)
axes[0, 0].set_xlabel('Economy (GDP per capita)', size = 15)
axes[0, 0].set_ylabel('Happiness Score', size = 15)

axes[0, 1].scatter(data2019['Social support'], data2019['Score'], marker ='o', color = 'g', s = 20)
axes[0, 1].set_xlabel('Social Support', size = 15)
axes[0, 1].set_ylabel('Happiness Score', size = 15)

axes[1, 0].scatter(data2019['Healthy life expectancy'], data2019['Score'], marker ='o', color = 'c', s = 20)
axes[1, 0].set_xlabel('Healthy Life Expectancy', size = 15)
axes[1, 0].set_ylabel('Happiness Score', size = 15)

axes[1, 1].scatter(data2019['Freedom to make life choices'], data2019['Score'], marker ='o', color = 'y', s = 20)
axes[1, 1].set_xlabel('Freedom to Make Life Choices', size = 15)
axes[1, 1].set_ylabel('Happiness Score', size = 15)

axes[2, 0].scatter(data2019['Generosity'], data2019['Score'], marker ='o', color = 'm', s = 20)
axes[2, 0].set_xlabel('Generosity', size = 15)
axes[2, 0].set_ylabel('Happiness Score', size = 15)

axes[2, 1].set_visible(False)

fig.text(.5, .08, 'Year 2019', ha='center', fontsize = 20)

plt.show()
```

**Hypothesis 3: I hypothesize that there is a negative correlation between perceptions of corruption is negatively correlated with happiness score from the year 2017–2019.**

```python
fig, axes = plt.subplots(2, 2, sharex = True, sharey = True, figsize = (20, 13))

axes[0, 0].scatter(data2017['Trust..Government.Corruption.'], data2017['Happiness.Score'], marker ='o', color = 'b', s = 20)
axes[0, 0].set_title('Year 2017', size = 15)

axes[0, 1].scatter(data2018['Perceptions of corruption'], data2018['Score'], marker ='o', color = 'c', s = 20)
axes[0, 1].set_title('Year 2018', size = 15)

axes[1, 0].scatter(data2019['Perceptions of corruption'], data2019['Score'], marker ='o', color = 'g', s = 20)
axes[1, 0].set_title('Year 2019', size = 15)

axes[1, 1].set_visible(False)

fig.text(.5, .08, 'Perceptions of Corruption', ha='center', fontsize = 20)
fig.text(.1, .45, 'Happiness Score', ha='center', rotation='vertical', fontsize = 20)

plt.show()
```

# Guiding Question 3: Does the world average happiness score steadily increase from the year 2015 to the year 2019?

**Hypothesis 4: I hypothesize that as time goes from 2015 to 2019, average happiness score increases.**

```python
happiness_averages = []
happiness_averages.append(data2015['Happiness Score'].mean())
happiness_averages.append(data2016['Happiness Score'].mean())
happiness_averages.append(data2017['Happiness.Score'].mean())
happiness_averages.append(data2018['Score'].mean())
happiness_averages.append(data2019['Score'].mean())

times = ['2015', '2016', '2017', '2018', '2019']

plt.plot(times, happiness_averages, marker = 'o', linewidth = 2, markersize = 5)

plt.xlabel('Time (Year)')
plt.ylabel('Average Happiness Score')

plt.title('Average Happiness Score in 2015-2019')

plt.show()
```

**Hypothesis 5: I hypothesize that two strong factors for happiness score also increase, from 2015–2019.**

```python
# gather averages from data in year 2015-2019
economy_averages = []
economy_averages.append(data2015['Economy (GDP per Capita)'].mean())
economy_averages.append(data2016['Economy (GDP per Capita)'].mean())
economy_averages.append(data2017['Economy..GDP.per.Capita.'].mean())
economy_averages.append(data2018['GDP per capita'].mean())
economy_averages.append(data2019['GDP per capita'].mean())
print("Economy Averages: ", economy_averages)
print()

healthy_life_expectancy_averages = []
healthy_life_expectancy_averages.append(data2015['Health (Life Expectancy)'].mean())
healthy_life_expectancy_averages.append(data2016['Health (Life Expectancy)'].mean())
healthy_life_expectancy_averages.append(data2017['Health..Life.Expectancy.'].mean())
healthy_life_expectancy_averages.append(data2018['Healthy life expectancy'].mean())
healthy_life_expectancy_averages.append(data2019['Healthy life expectancy'].mean())
print("Healthy Life Expectancy Averages: ", healthy_life_expectancy_averages)
print()

times = ['2015', '2016', '2017', '2018', '2019']

# Plot the data in two line plot
fig, axes = plt.subplots(1, 2, sharey = True, figsize = (15, 5))

axes[0].plot(times, economy_averages, marker = 'o', color = 'b', linewidth = 2, markersize = 5)
axes[0].set_xlabel('Time (Year)', size = 15)
axes[0].set_ylabel('Economy (GDP per capita)', size = 15)

axes[1].plot(times, economy_averages, marker = 'o', color = 'g', linewidth = 2, markersize = 5)
axes[1].set_xlabel('Time (Year)', size = 15)
axes[1].set_ylabel('Healthy Life Expectancy', size = 15)

fig.suptitle('Average Economy and Healthy Life Expectancy in 2015-2019', size = 20)

plt.show()
```
